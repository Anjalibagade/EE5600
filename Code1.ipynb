```python
import matplotlib.pyplot as plt
def nums(x1, x2, y1, y2, m, n):
    (x1, x2, y1, y2, m, n)=(-3,-8,-4,7,7,5)
 
    # Applying internal division formula
    
    a = (float)((m * x2)+(n * x1))/(m + n)
    b = (float)((m * y2)+(n * y1))/(m + n)
    
    # Applying external division formula
    
    p = (float)((m * x2)-(n * x1))/(m - n)
    q = (float)((m * y2)-(n * y1))/(m - n)
 
    # Printing result
    Internal_coordinate=(a,b)
    External_coordinate=(p,q)
    print (Internal_coordinate)
    print (External_coordinate)
 
    nums (x1, x2, y1, y2, m, n) 

# setting plot for internally divided coordinate

x=(-3,-71/12,-8)
y=(-4,29/12,7)
plt.plot(x,y)
plt.plot(x,y,'ro')
plt.xlabel('x - axis')
plt.ylabel('y-asis')
plt.title('Internally divided coordinate')
for i, j in zip(x, y):
 plt.text(i, j+0.5, '({}, {})'.format(i, j))
plt.show()

# setting plot for externally divided coordinate

x=(-3,-8,-41/2)
y=(-4,7,69/2)
plt.plot(x,y)
plt.plot(x,y,'go')
plt.xlabel('x - axis')
plt.ylabel('y-asis')
plt.title('Externally divided coordinate')
for i, j in zip(x, y):
 plt.text(i, j+0.5, '({}, {})'.format(i, j))
plt.show()
 


```

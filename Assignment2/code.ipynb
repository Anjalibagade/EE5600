```python

import matplotlib.pyplot as plt
x=(0,4)
y=(4,0)
plt.plot(x,y)
plt.plot(x,y,'ro')
plt.plot(0,4,'ro')
plt.plot(4,0,'ro')
plt.xlabel('x - axis')
plt.ylabel('y-asis')
for i, j in zip(x, y):
 plt.text(i, j+0.5, '({}, {})'.format(i, j))
plt.grid()

plt.show()
```
